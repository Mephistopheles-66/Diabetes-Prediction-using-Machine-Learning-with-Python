```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# For better visualization style
sns.set(style='whitegrid')
```

```python
# loading the diabetes dataset to a pandas DataFrame

diabetes_dataset = pd.read_csv('diabetes.csv')
```

# Distribution of Outcome

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', hue='Outcome', data=diabetes_dataset, palette='coolwarm', legend=False)
plt.title('Distribution of Diabetes Outcomes (0 = Non-Diabetic, 1 = Diabetic)')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show();
```

# Correlation Heatmap Shows relationships between all features

```python
plt.figure(figsize=(10,8))
correlation = diabetes_dataset.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show();
```

# Distribution of Key Features Visualize how glucose, BMI, and age differ by outcome.

```python
features = ['Glucose', 'BMI', 'Age']

plt.figure(figsize=(15,5))
for i, feature in enumerate(features):
    plt.subplot(1, 3, i+1)
    sns.histplot(data=diabetes_dataset, x=feature, hue='Outcome', kde=True, palette='coolwarm')
    plt.title(f'{feature} Distribution by Outcome')

plt.tight_layout()
plt.show();
```

# Shows pairwise relationships between features.

```python
sns.pairplot(diabetes_dataset[['Glucose', 'BMI', 'Age', 'Insulin', 'Outcome']], hue='Outcome', palette='coolwarm')
plt.suptitle('Pairwise Feature Relationships', y=1.02)
plt.show();
```

# Boxplots  Detect Outliers

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=diabetes_dataset[['Glucose', 'BMI', 'Age', 'Insulin']])
plt.title('Boxplot of Numerical Features (Outlier Detection)')
plt.show();
```

# Average Comparison by Outcome

```python
avg_outcome = diabetes_dataset.groupby('Outcome').mean(numeric_only=True)
avg_outcome.T.plot(kind='bar', figsize=(10,6), colormap='coolwarm')
plt.title('Average Feature Values by Outcome')
plt.ylabel('Mean Value')
plt.show();
```
